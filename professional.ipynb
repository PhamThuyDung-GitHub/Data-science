```python
import pandas as pd 
import numpy as np 
import os 
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates 
import matplotlib.ticker as mtick 
```

```python
df = pd.DataFrame([['Clothing', 0.1],
                   ['Home Decor', 0.43],
                   ['Electronics', 0.2],
                   ['Toys', 0.14]], columns = ['Spend_Category', 'Profit_Margin'])
df
```

```python
def percent_formatter(x, pos):
    return f"{x*100:.0f}%"
```

```python
fig, ax = plt.subplots(figsize=(12,9))
plt.bar(df['Spend_Category'], df['Profit_Margin'])
ax.yaxis.set_major_formatter(mtick.FuncFormatter(percent_formatter))

plt.grid(axis='y', alpha=0.45)
plt.title('Profit Margin for top five spend categories')
plt.ylabel('Profit Margin')
plt.xlabel('Category')

plt.show()
```

```python
df = df.sort_values("Profit_Margin", ascending=True, ignore_index=True)
df
```

```python
fig, ax = plt.subplots(figsize=(10,6))

title='Profit Margin for top five spend categories'
plt.text(s=title, 
         x=0, y=0.91,
         fontname='Times New Roman',
         fontweight='bold',
         fontsize=22,
         ha='left',
         transform=fig.transFigure)
subtitle = 'Consumers are willing to pay higher prices for decorative items for their home.'
plt.text(s=subtitle, 
         x=0, y=0.85,
         fontname='Times New Roman',
         fontsize=18,
         ha='left',
         transform=fig.transFigure)

bars = plt.barh(df['Spend_Category'],
                df['Profit_Margin'],
                color=['lightgray', 'lightgray', 'lightgray', 'lightgray', 'black'],
                edgecolor='none')

plt.margins(y=0.14)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.set_xticklabels([])
ax.set_xticks([])

ax.tick_params(axis='y', left=False)
plt.yticks(fontsize=14, fontweight='bold')

plt.bar_label(
    bars,
    labels=[f'{x:.0%}' for x in bars.datavalues],
    padding=12,
    fontsize=15
)

ax.plot(
    [-0.025, .94],
    [.8335, .8335],
    transform = fig.transFigure,
    clip_on = False,
    color = 'tab:blue',
    linewidth=3
)

ax.add_patch(
    plt.Rectangle(

        (-0.03, .95),
        0.015,
        -0.12,
        facecolor='tab:blue',
        transform=fig.transFigure,
        clip_on=False,
        linewidth=0
    )
)

footnote="*Data from Jan-2022 to Mar-2023"
plt.text(s=footnote, 
         x=0,
         y=0.13,
         fontname='Arial',
         fontstyle='italic',
         fontsize=15,
         ha='left',
         transform=fig.transFigure);
```
