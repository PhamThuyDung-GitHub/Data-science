```python
import duckdb 
import pandas as pd 
```

```python
%load_ext sql 
```

```python
conn = duckdb.connect()
```

```python
%sql conn --alias duckdb
```

```python
data = [["ğŸğŸ", "NY", 10],
        ["ğŸŠğŸŠ", "CA", 15],
        ["ğŸğŸ", "CA", 20],
        ["ğŸğŸ", "NY", 18]]

data = pd.DataFrame(data, columns=["Fruit", "Store", "Amount"])
data
```

```python
query = """ 
SELECT Fruit, Store, Sum(Amount)
FROM data 
GROUP BY GROUPING SETS (
(Fruit), (Store)
);
"""

duckdb.sql(query).df()
```

```python
query = """
SELECT Fruit, Store, Sum(Amount)
FROM data 
GROUP BY ROLLUP (
Fruit, Store);
"""

duckdb.sql(query).df()
```

```python
query = """
SELECT Fruit, Store, Sum(Amount)
FROM data 
GROUP BY CUBE (
Fruit, Store)
"""

duckdb.sql(query).df()
```
